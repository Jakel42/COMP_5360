```python
import os
import glob
import pandas as pd
import numpy as np
main_directory = os.path.normpath(os.getcwd() + os.sep + os.pardir)
csv_directory = os.path.join(main_directory, 'data_import')
```

```python
player_stat_df = pd.read_csv(os.path.join(csv_directory, 'player_stat.csv'))
player_stat_df.head(10)
```

```python
player_stat_df.columns
```

```python
#lets make the hedaer into a more meaningful description

name_map = {'Rk':'Rank','Pos':'Position','Tm':'Team','G':'Games','GS':'Games started','MP':'Minutes played per game'
             ,'FG':'Field goals per game','FGA':'Field goals attempt per game','FG%':'Field goal percent','3P':'3 point field goal per game',
             '3PA':'3 point field goal attempt per game', '3P%':'3 point field goal percentage','2P':'2 point field goal per game', '2PA':'2 point field goal attempt per game',
             '2P%':'2 point field goal percentage', 'eFG%':'Effective field goal percentage', 'FT':'Free throws per game',
             'FTA':'Free throw attempt per game','FT%':'Free throw percentage','ORB':'Offensive rebounds per game','DRB':'Defensive rebounds per game',
             'TRB':'Total rebounds per game','AST':'Assist per game','STL':'Steals per game','BLK':'Blocks per game','TOV':'Turn overs per game',
             'PF':'Personal fouls per game','PTS':'Points per game'}

player_stat_df = player_stat_df.rename(columns=name_map)
```

```python
player_stat_df.isnull().sum()
```

```python
pd.set_option('display.max_columns', None)
player_stat_df.head(20)
```

```python
#Looks like some of these guys are null due to having 0 shots attempt to 0 shot made ratio, percent is NA
#Just in case lets check if that is true
player_stat_df.columns
```

```python
result = player_stat_df['Field goals per game'] / player_stat_df['Field goals attempt per game']
result = result.replace([np.inf, -np.inf, np.nan], 0)
player_stat_df['Field goal percent'] = result

result = player_stat_df['3 point field goal per game'] / player_stat_df['3 point field goal attempt per game']
result = result.replace([np.inf, -np.inf, np.nan], 0)
player_stat_df['3 point field goal percentage'] = result

result = player_stat_df['2 point field goal per game'] / player_stat_df['2 point field goal attempt per game']
result = result.replace([np.inf, -np.inf, np.nan], 0)
player_stat_df['2 point field goal percentage'] = result

result = player_stat_df['Free throws per game'] / player_stat_df['Free throw attempt per game']
result = result.replace([np.inf, -np.inf, np.nan], 0)
player_stat_df['Free throw percentage'] = result

result = (player_stat_df['Field goals per game'] + 0.5 *  player_stat_df['3 point field goal per game'])/ player_stat_df['Field goals attempt per game']
result = result.replace([np.inf, -np.inf, np.nan], 0)
player_stat_df['Effective field goal percentage'] = result
```

```python
player_stat_df.to_csv('cleaned_player_stat.csv')
```
