```python
import os
import glob
import pandas as pd
import numpy as np
main_directory = os.path.normpath(os.getcwd() + os.sep + os.pardir)
csv_directory = os.path.join(main_directory, 'data_import')

def_df = pd.read_csv(os.path.join(csv_directory, 'team_adv.csv'))
pd.set_option('display.max_columns', None)

def_df.head(10)
```

```python
def_df.columns
```

```python
def_df = def_df.drop(columns=['Unnamed: 17', 'Unnamed: 22','Unnamed: 27'])
```

```python
name_mapping = {
    'Rk': 'Rank',
    'Teamâ–²': 'Team',
    'Age': 'Player Age',
    'W': 'Wins',
    'L': 'Losses',
    'PW': 'Pythagorean Wins',
    'PL': 'Pythagorean Losses',
    'MOV': 'Margin of Victory',
    'SOS': 'Strength of Schedule',
    'SRS': 'Simple Rating System',
    'ORtg': 'Offensive Rating',
    'DRtg': 'Defensive Rating',
    'NRtg': 'Net Rating',
    'Pace': 'Pace Factor',
    'FTr': 'Free Throw Rate',
    '3PAr': 'Three-Point Attempt Rate',
    'TS%': 'True Shooting Percentage',
    'eFG%': 'Effective Field Goal Percentage',
    'TOV%': 'Turnover Percentage',
    'ORB%': 'Offensive Rebound Percentage',
    'FT/FGA': 'Free Throws per Field Goal Attempt',
    'eFG%.1': 'Opponent Effective Field Goal Percentage',
    'TOV%.1': 'Opponent Turnover Percentage',
    'DRB%': 'Defensive Rebound Percentage',
    'FT/FGA.1': 'Opponent Free Throws per Field Goal Attempt',
    'Arena': 'Arena',
    'Attend.': 'Attendance',
    'Attend./G': 'Attendance per Game'
}

def_df = def_df.rename(columns=name_mapping)
def_df
```

```python
def_df.isnull().sum()
```

```python
data = def_df[def_df['Rank'].isna()]
data
```

```python
#looks like some team are rankless with no win or lose recorded
def_df = def_df[def_df['Rank'].notna()]
```

```python
def_df.isnull().sum()
```

```python
def_df['Team'] = def_df['Team'].str.replace('*', '')
def_df['Team'].unique()
```

```python
attendance_median = def_df['Attendance'].median()

# Replace NaN values with the median
def_df['Attendance'].fillna(attendance_median, inplace=True)
def_df.to_csv('cleaned_defensive.csv')
```
