```python
import os
import glob
import pandas as pd
import re

main_directory = os.path.normpath(os.getcwd() + os.sep + os.pardir)
data_directory = os.path.join(main_directory, 'TeamStats')
file_names = [f for f in os.listdir(data_directory) if os.path.isfile(os.path.join(data_directory, f))]
file_names
```

```python
#lets read in 2000 def, adv,off, and playoffs


df_df = pd.read_csv(os.path.join(data_directory, '2000Def.csv'))
df_adv = pd.read_csv(os.path.join(data_directory, '2000Adv.csv'),header=1)
df_off = pd.read_csv(os.path.join(data_directory, '2000Off.csv'))
df_play = pd.read_csv(os.path.join(data_directory, '2000Playoffs.csv'),header=1)
```

```python
#seperate file names
off_files = [file for file in file_names if 'Off.csv' in file]
def_files = [file for file in file_names if 'Def.csv' in file]
adv_files = [file for file in file_names if 'Adv.csv' in file]
playoff_files = [file for file in file_names if 'Playoffs.csv' in file]

off_files = sorted(off_files)
def_files = sorted(def_files)
adv_files = sorted(adv_files)
playoff_files = sorted(playoff_files)

#initalize column headers and check if all file format is the same
off_header = df_off.columns
def_header = df_df.columns
adv_header = df_adv.columns
play_header = df_play.columns
off_header
```

```python
#initalize some dataframe
df_offensive = pd.DataFrame()
df_defensive = pd.DataFrame()
df_playoff = pd.DataFrame()
df_advance = pd.DataFrame()


for i in range(len(off_files)):
    df = pd.read_csv(os.path.join(data_directory, off_files[i]))
    if df.columns.equals(off_header):
        df_offensive = pd.concat([df_offensive,df])
    else:
        print('this file header didnt match: ',off_files[i])
    
    df = pd.read_csv(os.path.join(data_directory, def_files[i]))
    if df.columns.equals(def_header):
        df_defensive = pd.concat([df_defensive,df])
    else:
        print('this file header didnt match: ',def_files[i])
        
    df = pd.read_csv(os.path.join(data_directory, adv_files[i]),header=1)
    if df.columns.equals(adv_header):
        df_advance = pd.concat([df_advance,df])
    else:
        print('this file header didnt match: ',adv_files[i])
        
    df = pd.read_csv(os.path.join(data_directory, playoff_files[i]),header=1)
    if df.columns.equals(play_header):
        df_playoff = pd.concat([df_playoff,df])
    else:
        print('this file header didnt match: ',playoff_files[i])
```
